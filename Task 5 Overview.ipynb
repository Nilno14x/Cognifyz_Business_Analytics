# Task 5: Reasons for Investment

## Analyze and summarize reasons for investment choices

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Data_set 2 - Copy.csv"
df = pd.read_csv(file_path)
```

```python
reason_columns = df[['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']]
```

```python
reasons = reason_columns.stack().reset_index(drop=True)
```

```python
reason_counts = reasons.value_counts()
```

```python
print("Summary of Investment Reasons:\n", reason_counts)
```

```python
# Plotting a horizontal bar chart
plt.figure(figsize=(10, 6))
reason_counts.sort_values().plot(kind='barh', color='lightcoral')

# Adding labels and title
plt.title('Distribution of Investment Reasons (Horizontal Bar Chart)', fontsize=14)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Investment Reasons', fontsize=12)
plt.tight_layout()
plt.show()
```

### Summary:


1. Most Common Reasons: "Capital Appreciation" is the most common reason, appearing 30 times across different columns.
2. Other Frequent Reasons: "Assured Returns", "Better Returns", and "Risk Free" are also frequently cited reasons.
3. Unique or Less Frequent Reasons: "Tax incentives", "Liquidity", "High Interest Rates", and "Tax Benefits" are mentioned less often, indicating they may be secondary considerations.
