 ## Calculate the average investment duration.

```python
import pandas as pd
```

```python
file_path = "Data_set 2 - Copy.csv"
df = pd.read_csv(file_path)
```

Since the Duration column in the dataset containig th ecatregorical value. So, in this case we have to handel it by converting into numerical values for meaningful analysis.

```python
duration_format = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}
```

```python
df['duration_numeric'] = df['Duration'].map(duration_format)
```

```python
df['duration_numeric']
```

```python
print("\nDataset with Numeric Durations:\n")
df
```

```python
average_duration = df['duration_numeric'].mean()
```

```python
print(f"The Average Duration of Investment : {average_duration:.2f} years")
```
