# Task 10 : Correlation Analysis

## Explore potential correlations between factors.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
file_path = "Data_set 2 - Copy.csv"
df = pd.read_csv(file_path)
```

```python
# Convert "Expected Returns" to numerical values (taking the middle value of the range)
def convert_returns(return_range):
    low, high = return_range.replace('%', '').split('-')
    return (int(low) + int(high)) / 2
```

```python
df['expect_num'] = df['Expect'].apply(convert_returns)
```

```python
df['expect_num']
```

```python
correlation_matrix = df[['age', 'duration_numeric', 'expect_num']].corr()
```

```python
print("Correlation Matrix:\n", correlation_matrix)
```

```python
# Visualize the correlation using a heatmap
plt.figure(figsize=(8, 6))
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix between Age, Investment Duration, and Expected Returns')
plt.tight_layout()
plt.show()
```

```python
# Scatter plots to visualize relationships
plt.figure(figsize=(12, 6))

# Scatter plot: Age vs. Investment Duration
plt.subplot(1, 2, 1)
plt.scatter(df['age'], df['duration_numeric'], color='skyblue')
plt.title('Age vs. Investment duration')
plt.xlabel('Age')
plt.ylabel('Investment Duration (Years)')
plt.grid(True)

# Scatter plot: Age vs. Expected Returns
plt.subplot(1, 2, 2)
plt.scatter(df['age'], df['expect_num'], color='salmon')
plt.title('Age vs. Expected returns')
plt.xlabel('Age')
plt.ylabel('Expected Returns (%)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 1. Pair Plot to visualize relationships between age, investment duration, and expected returns
sb.pairplot(df[['age', 'duration_numeric', 'expect_num']], kind='scatter', hue='age')
plt.suptitle('Pair Plot: Relationships between Age, Investment Duration, and Expected Returns', y=1.02)
plt.tight_layout()
plt.show()
```
