# Task 2 : Gender Distribution


## Visualize gender distribution in the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Data_set 2 - Copy.csv"
df = pd.read_csv(file_path)
```

```python
Gender = df["gender"]
Gender.count()
```

```python
Gender_counts = Gender.value_counts()

print("Gender Distribution:\n", Gender_counts)
```

```python
plt.figure(figsize=(6, 4))
Gender_counts.plot(kind='bar', color=['#1f77b4', '#ff7f0e'])
plt.title('Gender Distribution (Bar Chart)', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
Gender_counts.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['#1f77b4', '#ff7f0e'], 
    labels=Gender_counts.index
)
plt.title('Gender Distribution (Pie Chart)', fontsize=14)
plt.ylabel('')  # Remove y-axis label for a cleaner look
plt.show()
```
