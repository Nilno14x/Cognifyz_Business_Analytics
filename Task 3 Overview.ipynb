# Task 3 : Descriptive Statistics

## Present basic statistics for numerical columns

```python
import pandas as pd
file_path = "Data_set 2 - Copy.csv"
df = pd.read_csv(file_path)
```

```python
numerical_columns = df.select_dtypes(include=['number'])
numerical_columns
```

```python
for column in numerical_columns.columns:
    mean_value = numerical_columns[column].mean()
    median_value = numerical_columns[column].median()
    std_value = numerical_columns[column].std()
        
    print(f"\nStatistics for '{column}' \n Mean: {mean_value} \n Median: {median_value} \n Standard Deviation: {std_value.round(4)}")
    
```
