# Task 4: Most Preferred Investment Avenue


## Identify the most preferred investment avenue

```python
import pandas as pd
file_path = "Data_set 2 - Copy.csv"
df = pd.read_csv(file_path)
```

```python
Analyze_column = df[['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']]
```

```python
Analyze_column
```

```python
for column in Analyze_column.columns:
    frequency = df[column].value_counts()
    
    # Get the most frequent value and its count
    most_frequent_value = frequency.idxmax()
    highest_frequency = frequency.max()
    
    print(f"Column: {column}")
    print(f"Most frequent value: {most_frequent_value}, Frequency: {highest_frequency}")
    print('-' * 50)
```
