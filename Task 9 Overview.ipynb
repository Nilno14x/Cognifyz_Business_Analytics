#  Task 9 : Expectations from Investments

## Summarize participant's expectations from investments.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Data_set 2 - Copy.csv"
df = pd.read_csv(file_path)
```

```python
Expect = df["Expect"]
```

```python
Expect
```

```python
Expect.value_counts()
```

```python
expectation_description = {
    '30%-40%': "This group wants higher returns, meaning they’re willing to take on more risk for bigger rewards.",
    '20%-30%': "Participants expect moderate returns, usually from high-risk investments like stocks or mutual funds.",
    '10%-20%': "These participants expect lower returns, suggesting they prefer safer investments like bonds or fixed deposits."
}
```

```python
for expectation, description in expectation_description.items():
    print(f"\nExpectation: {expectation} \nDescription: {description}")
```

```python
ex_p = Expect.value_counts()

plt.figure(figsize=(7, 7))
plt.pie(ex_p , labels=ex_p.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Investment Expectations')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.

plt.show()
```
